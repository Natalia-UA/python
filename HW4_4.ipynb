```python
import pandas as pd

df = pd.read_csv(r'C:\Users\Natalia\Downloads\stack-overflow-developer-survey-2023\survey_results_public.csv')

print(df)
```

# Скільки респондентів пройшли опитування?

```python
print('Number of respondents: ',df['ResponseId'].count())
```

# Скільки респондентів відповіли на всі запитання?

```python
df.dropna(inplace=True)
print('Number of respondents who answered all questions: ',df['ResponseId'].count())
```

# Які значення мір центральної тенденції для досвіду (WorkExp) респондентів?

```python
df.groupby('WorkExp') ['ConvertedCompYearly'] .describe()
```

# Скільки респондентів працює віддалено?

```python
rem = df.loc[df['RemoteWork'] == 'Remote']

print('Works remotely: ', rem['RemoteWork'].count())
```

# Який відсоток респондентів програмує на Python?

```python
lang=df.loc[df['LanguageHaveWorkedWith']=='Python']
py=lang['ResponseId'].count()
all=df['ResponseId'].count()
percent=round(py/all*100,2)
print('Programming in Python: ', percent,'%')
```

# Скільки респондентів навчалося програмувати за допомогою онлайн курсів?

```python
onl=df.loc[(df['LearnCodeOnline'] != 'NA') & (df['LearnCodeOnline'] != '')]
print('Learned to program on-line: ', onl['ResponseId'].count())
```

# Серед респондентів що програмують на Python в групуванні по країнам, яка середня та медіанна сума компенсації (ConvertedCompYearly) в кожній країні?

```python
#медіана
lang=df.loc[df['LanguageHaveWorkedWith']=='Python']
lang.groupby('Country')['ConvertedCompYearly'].median()
```

```python
#середнє
lang=df.loc[df['LanguageHaveWorkedWith']=='Python']
round(lang.groupby('Country')['ConvertedCompYearly'].mean(),2)
```

# Які рівні освіти мають 5 респондентів з найбільшою компенсацією?

```python
top=df.sort_values(by = 'ConvertedCompYearly', ascending = False).head(5)
value = top.iloc[[0,1,2,3,4],[7]]
print(value)
```

# В кожній віковій категорії, який відсоток респондентів програмує на Python?

```python
#загальна кількість респондентів у групі
total=df.groupby('Age')   ['LanguageHaveWorkedWith'].count()

#кількість респондентів у групі, що програмує на Python
lang=df.loc[df['LanguageHaveWorkedWith']=='Python']
p=lang.groupby('Age')  ['LanguageHaveWorkedWith'].count()


per_py=round(p/total *100,2)
print(per_py)
```

# Серед респондентів що знаходяться в 75 перцентилі за компенсацією середнього і працюють віддалено, які індустрії є найрозповсюдженішими?

```python
q=df['ConvertedCompYearly'].quantile(0.75)
df1=df.loc[(df['RemoteWork'] == 'Remote') & (df['ConvertedCompYearly'] >= q)]

print('The most common industry: ', df1['Industry'].mode())

```
